<h3>3.1 Descubrimiento y entendimiento de datos </h3>
<p align = "justify"> La base de datos que tenemos como insumo para analizar e implementar modelos predictivos, contiene 2 columnas con las variables **Fecha** y **Unidades** las cuales son muy dicientes, y pueden ser usadas, especialmente la primer variable mencionada, para la creación y reclasificación de nuevas variables que puedan tener efectos significativos sobre la variable respuesta, que en este caso es el número de registros diarios.</p>

La estructura de las variables que contiene los datos es la siguiente:

```python
import pandas as pd
import datetime

df = pd.read_excel("/content/registros_autos_entrenamiento(1).xlsx")
```

```python
df['Año'] = df['Fecha'].dt.year 
df['Mes'] = df['Fecha'].dt.month
df['Dia'] = df['Fecha'].dt.day 
df=df[['Año','Mes','Dia','Unidades']]
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,7))
Unidades_año=df.groupby(["Mes","Año"]).sum()["Unidades"].unstack().plot(ax=ax)
ax.grid()
```

Se puede observar que cada mes es variable en su icremento de inscripción de vehiculos, por tanto a simple inspección no se puede denotar como un sistema lineal variable en el tiempo, graficamente se aporta que cada año que pasa del 2012 a 2017 tiende a disminuir el numero de vehiculos inscritos ecxeptuando el año 2013 y 2014, siendo este ultimo mayor ante su antecesor, ademas de que en el ultimo mes de cada uno de los años se reportan el mayor ingreso de vehiculos al registro al sistema de transito nacional, esto puede ser debido a la fiestas dicembrinas del pais y su cultura de compra en estas fechas en el pais colombiano.
